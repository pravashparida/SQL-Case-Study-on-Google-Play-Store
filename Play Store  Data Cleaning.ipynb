```python
import pandas as pd
import numpy as np
import datetime
```

```python
data=pd.read_csv('googleplaystore(impure).csv')
```

```python
# checking null values
data.isnull().sum()
```

```python
# dropping null values
data.dropna(inplace=True)
```

```python
data.head()
```

```python
# dropping unnecessary columns
data.drop(columns='Unnamed: 0',inplace=True)
```

```python
data
```

```python
# changing data time to desired format
data['Last Updated']=pd.to_datetime(data['Last Updated'],errors='coerce')
```

```python
data.info()
```

```python
data['Last Updated']=data['Last Updated'].dt.strftime('%Y-%m-%d')
```

```python
data['Last Updated']=pd.to_datetime(data['Last Updated'],errors='coerce')
```

```python
data.info()
```

```python
# cleaning install columns
data['Installs']=data['Installs'].str.replace('+','')
data['Installs']=data['Installs'].str.replace(',','')
```

```python
data['Installs']=data['Installs'].astype(int)
```

```python
data
```

```python
# cleaning reviews column
data.drop(data[data['Reviews']=='3.0M'].index,inplace=True)
```

```python
data['Reviews']=data['Reviews'].astype(int)
```

```python
data
```

```python
data.info()
```

```python
# cleaning price column
data['Price']=data['Price'].str.replace('$','')
data['Price']=data['Price'].astype(float)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# now our df is cleaned we will dump it
data.to_csv('playstore.csv',index=False)
```
